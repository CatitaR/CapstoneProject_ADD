```python
import pandas as pd
import numpy as np
import zipfile
import wget
import requests
import io
import regex as re
import os
```

```python
defunciones = pd.read_csv("https://raw.githubusercontent.com/CatitaR/CapstoneProject_ADD/main/data/raw/defunciones_deis_2017.csv", 
                          encoding='latin-1',
                          sep=";")
```

```python
defunciones
```

```python
defunciones = defunciones[(defunciones['CODIGO_COMUNA_RESIDENCIA']!=99999) &
                          (defunciones['GLOSA_CAPITULO_DIAG1']=='Traumatismos, envenenamientos y algunas otras consecuencias de causas externas')]
defunciones = defunciones[['CODIGO_COMUNA_RESIDENCIA','GLOSA_CAPITULO_DIAG1','FECHA_DEF']]
def_2017 = (defunciones.groupby(by=['CODIGO_COMUNA_RESIDENCIA','GLOSA_CAPITULO_DIAG1'])
               .count().reset_index()
               .rename(columns = {'CODIGO_COMUNA_RESIDENCIA':'Comuna',
                                  'GLOSA_CAPITULO_DIAG1':'causa',
                                  'FECHA_DEF':'n_def'}))
```

```python
defunciones['GLOSA_CAPITULO_DIAG1'].value_counts()
```

```python
def_2017
```

```python
proy_pob = pd.read_excel("https://github.com/CatitaR/CapstoneProject_ADD/blob/main/data/raw/estimaciones-y-proyecciones-2002-2035-comuna-y-%C3%A1rea-urbana-y-rural.xlsx?raw=true")
```

```python
proy_pob
```

```python
proy_pob_2017 = proy_pob[['Comuna','Nombre Comuna','Poblacion 2017']].groupby(['Comuna','Nombre Comuna']).sum('Poblacion 2017').reset_index()
```

```python
proy_pob_2017
```

```python
tasa_def = def_2017.merge(proy_pob_2017,on='Comuna',how='outer')
tasa_def['tasa_def'] = tasa_def['n_def']/tasa_def['Poblacion 2017']*100000
tasa_def["tasa_def"].fillna(0, inplace = True)
```

```python
tasa_def
```

```python
pob_rural = (proy_pob[['Comuna','Poblacion 2017','Area (1=Urbano 2=Rural)']]
                .groupby(['Comuna','Area (1=Urbano 2=Rural)']).sum().reset_index())
pob_rural = pob_rural[pob_rural['Area (1=Urbano 2=Rural)']==2]
pob_rural = pob_rural[['Comuna','Poblacion 2017']].rename(columns = {'Poblacion 2017': 'pob_rural'})
pob_rural = pob_rural.merge(proy_pob_2017,on='Comuna',how='outer')
pob_rural['prop_rural'] = pob_rural['pob_rural']/pob_rural['Poblacion 2017']*100
pob_rural = pd.DataFrame(pob_rural[['Comuna','prop_rural']])
```

```python
pob_rural
```

```python
tasa_def = tasa_def.merge(pob_rural,on='Comuna',how='outer')
tasa_def
```

```python
comunas = tasa_def[['Nombre Comuna','Comuna']].set_index('Nombre Comuna')
comunas_dict = comunas.to_dict()['Comuna']
```

```python
comunas_dict
```

```python
r = requests.get('https://github.com/CatitaR/CapstoneProject_ADD/blob/main/data/raw/casen2017.zip?raw=true')
z = zipfile.ZipFile(io.BytesIO(r.content))
```

```python
casen_2017 = pd.read_stata(z.open('Casen 2017.dta'),columns = ['folio','o','id_vivienda', 'hogar', 'region', 'provincia', 'comuna','zona', 'expr', 'expc',
                        'yautcor','ytotcor','yautcorh','ypc','pobreza','r3'])
```

```python
casen_2017
```

```python
casen_2017.region.value_counts()
```

```python
casen_2017['comuna'] = np.where(casen_2017['comuna']=='Los Ángeles','Los Angeles',
                               np.where(casen_2017['comuna']=='Pitrufquén','Pitrufquen',casen_2017['comuna']))
casen_2017['Comuna'] = casen_2017['comuna'].map(comunas_dict)
casen_2017['etnia'] = np.where(casen_2017['r3']=='No pertenece a ningún pueblo indígena',0,
                                         np.where(casen_2017['r3']=='No sabe/no responde',0,1))
```

```python
casen_2017['Comuna'] = casen_2017['Comuna'].astype(int)
```

```python
casen_hogar = casen_2017.drop_duplicates(casen_2017.columns[~casen_2017.columns.isin(['folio'])],
                        keep='first')
```

```python
def weighted_average(df, values, weights):
    return sum(df[weights] * df[values]) / df[weights].sum()
```

```python
ing = (pd.DataFrame(casen_2017.groupby('Comuna').apply(weighted_average, 'ypc', 'expc'))
       .reset_index()
      .rename(columns={0:'ingreso'}))
```

```python
ing
```

```python
etnia = (pd.DataFrame(casen_2017.groupby('Comuna').apply(weighted_average, 'etnia', 'expc'))
         .reset_index()
        .rename(columns={0:'etnia'}))
etnia['etnia'] = etnia['etnia']*100
```

```python
etnia
```

```python
tasa_def = tasa_def.merge(etnia,on='Comuna',how='outer').merge(ing,on='Comuna',how='outer')
```

```python
tasa_def[tasa_def['ingreso'].isnull()]
#las siguientes comunas no están en casen 2017:
#Isla de Pascua, Cochamó, Chaitén, Futaleufú, Hualaihué, Primavera, Colchane, Ollagüe, Juan Fernández, Palena
#Lago Verde, Guaitecas, O'Higgins, Tortel, Laguna Blanca, Río Verde, San Gregorio, Cabo de Hornos
#Antártica, Timaukel, Torres del Paine, General Lagos
```

```python
tasa_def = tasa_def.dropna(subset=['etnia','ingreso'])
```

```python
denuncias = pd.read_excel('https://github.com/CatitaR/CapstoneProject_ADD/blob/main/data/raw/datos%20denuncias%202017.xls?raw=true')
```

```python
denuncias
```

```python
denuncias['glosa_comuna'] = denuncias['Unidad territorial'].replace(r"\(Región de Arica y Parinacota\)","",regex=True)
denuncias['glosa_comuna'] = denuncias['glosa_comuna'].replace(r"Los Ángeles","Los Angeles",regex=True)
denuncias['glosa_comuna'] = denuncias['glosa_comuna'].replace(r"Pitrufquén","Pitrufquen",regex=True)
denuncias['glosa_comuna'] = denuncias['glosa_comuna'].replace(r"O’higgins","O'Higgins",regex=True)
denuncias['glosa_comuna'] = denuncias['glosa_comuna'].replace(r"\(Región de los Ríos\)","",regex=True)
denuncias['glosa_comuna'] = denuncias['glosa_comuna'].replace(r"\s+$","",regex=True)
denuncias['glosa_comuna'] = denuncias['glosa_comuna'].replace(r" De"," de",regex=True)
denuncias['glosa_comuna'] = denuncias['glosa_comuna'].replace(r" La "," la ",regex=True)
denuncias['Comuna'] = denuncias['glosa_comuna'].map(comunas_dict)
```

```python
denuncias[denuncias['Comuna'].isnull()].glosa_comuna
```

```python
denuncias.columns = denuncias.columns.str.replace(' 2017','tasa_denuncias')
denuncias = denuncias[['Comuna','tasa_denuncias']]
```

```python
denuncias
```

```python
tasa_def = tasa_def.merge(denuncias,on='Comuna',how='outer')
```

```python
tasa_def[tasa_def['tasa_denuncias'].isnull()]
#no hay comunas que estén en la base de datos de denuncias pero no en defunciones
```

```python
tasa_def = tasa_def.dropna(subset=['etnia','ingreso'])
```

```python
tasa_def['region'] = np.where(tasa_def['Comuna']>=10000, tasa_def['Comuna'].astype(str).str[0:2], tasa_def['Comuna'].astype(str).str[0])
tasa_def['region'] = tasa_def['region'].astype('int64')
tasa_def['rm'] = np.where(tasa_def['region']==13, 1,0)
```

```python
tasa_def_rm = tasa_def[tasa_def['region']==13]
tasa_def_otras = tasa_def[tasa_def['region']!=13]
```
